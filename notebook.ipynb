```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
df = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
df
```

```python
gender_encoder = LabelEncoder()
df['Gender'] = gender_encoder.fit_transform(df['Gender'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
geo_encoder = ohe.fit_transform(df[['Geography']])
geo_encoder
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geo_encoder.toarray()
```

```python
encoded_geo = pd.DataFrame(geo_encoder.toarray(), columns=ohe.get_feature_names_out(['Geography']))
encoded_geo
```

```python
df = pd.concat([df.drop('Geography',axis=1),encoded_geo],axis=1)
df
```

```python
# Saving pickle files for encoders
# with open('label_ecoder_gender.pkl','wb') as file:
#     pickle.dump(LabelEncoder(),file)

with open('ohe_geography','wb') as file:
    pickle.dump(ohe,file)
```

```python
with open('label_ecoder_gender.pkl','wb') as file:
    pickle.dump(gender_encoder,file)
```

```python
df.head()
```

```python
X = df.drop('Exited', axis=1)
Y = df['Exited']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

# Scaling

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape[1]
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
## Model Building

model = Sequential(
    [
        Dense(64,activation='relu', input_shape=(X_train_scaled.shape[1],)),
        Dense(32,activation='relu'),
        Dense(1, activation='sigmoid')
    ]
)
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.1)
model.compile(optimizer=opt, loss='binary_crossentropy',metrics= ['accuracy'])
```

```python
log_directory = "logs/fit" + datetime.datetime.now().strftime("%d/%m/%Y -- %H%M%S")
tf_callback = TensorBoard(log_dir= log_directory, histogram_freq=1)
```

```python
early_stop = EarlyStopping(monitor='val_loss', verbose= 1, patience= 10, restore_best_weights= True)
```

```python
history = model.fit(X_train_scaled, Y_train, validation_data=(X_test_scaled, Y_test), epochs=100, callbacks=[tf_callback, early_stop])
```

```python
model.save('model.h5')
```
